```python
n = 4
A = 65
def node_name(i, j):
    return f'N{i}{j}'

df['letters'] = ''

def adding_letters(column, ch, i, j):
    for ind in df.index:
        if(df[column][ind] == f'N{i}{j}'):
            c = df['letters'][ind]
            df.loc[ind, 'letters'] = c + ch

def giving_loc(i, j):
    adding_letters('node1', ch, i, j)
    adding_letters('node2', ch, i, j)
    
reverse = 0
for c in range(ord('A'), ord('A') + 4 * (n-1)):
    ch = chr(c)
    # print(ch, c - A)
    c = c - A
    # A B C
    # D E F
    # G H I
    # J K L
    '''
    Lets have 4 groups of letters
    '''
    if(int(c/3) == 0):
        giving_loc(0, c)
        # print(0, c)
    if(int(c/3) == 1):
        giving_loc(c - n + 1, 3)
        # print(c - n + 1, 3)
    if(int(c/3) == 2):
        giving_loc(3, c - n + 1 + reverse)
        # print(3, c - n + 1 + reverse)
        reverse -= 2
        # print(reverse)
    if(int(c/3) == 3):
        giving_loc(c + reverse, 0)
        # print(c + reverse, 0, c)
        reverse -= 2
        # print(reverse)
df
```

```python
for ind in df.index:
    if(len(df['letters'][ind]) < 2):
        df = df.drop([ind]);
df
```

```python
df1 = df
def swap_rows(df, row1, row2):
    df.iloc[row1], df.iloc[row2] =  df.iloc[row2].copy(), df.iloc[row1].copy()
    return df

df1 = swap_rows(df1, 1, 11)
df1 = swap_rows(df1, 2, 11)
df1 = swap_rows(df1, 3, 8)
df1 = swap_rows(df1, 4, 11)
df1 = swap_rows(df1, 7, 10)
df1 = swap_rows(df1, 5, 7)
df1
```
