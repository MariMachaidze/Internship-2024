```python
ch = 'A'
n = 4
def node_name(i, j):
    return f'N{i}{j}'

df_compare['letters'] = ''

def adding_letters(column, ch, i, j):
    for ind in df_compare.index:
        if(df_compare[column][ind] == f'N{i}{j}'):
            c = df_compare['letters'][ind]
            df_compare.loc[ind, 'letters'] = c + ch

for j in range(n):
    adding_letters('node1', ch, 0, j)
    adding_letters('node2', ch, 0, j)
    ch = chr(ord(ch) + 1)
    print(0, j)

for i in range(1, n):
    adding_letters('node1', ch, i, 3)
    adding_letters('node2', ch, i, 3)
    ch = chr(ord(ch) + 1)
    print(i, 3)

for j in range(n-2, -1, -1):
    adding_letters('node1', ch, 3, j)
    adding_letters('node2', ch, 3, j)
    ch = chr(ord(ch) + 1)
    print(3, j)

for i in range(n-2, 0, -1):
    adding_letters('node1', ch, i, 0)
    adding_letters('node2', ch, i, 0)
    ch = chr(ord(ch) + 1)
    print(i, 0)

df_compare
```
