```python
# Function to get node name
def node_name(i, j):
    return f'N{i}{j}'
```

```python
n = 4
ch = 'A'

data = []
def loop(ib,ie,ir,jb,je,jr):
    global ch
    for i in range(ib, ie, ir):
        for j in range(jb, je, jr):
            node = node_name(i, j)
            print('Node {}: {}'.format(node, ch)) 
            data.append([node, ch])
            ch = chr(ord(ch) + 1)
            
loop(0, 1, 1, 0, n-1, 1) # top line
print()
loop(0, n-1, 1, n-1, n, 1) # right line
print()
loop(n-1, n, 1, n-1, 0, -1) # bottom line
print()
loop(n-1, 0, -1, 0, 1, 1) # right line

df_letters = pd.DataFrame(data, columns=['node', 'letters'])
df_letters
```

```python
df = pd.merge(df, df_letters, how="inner", on=["node"])
df
```

```python
n = 4

data = []

def each_step(node1, node2):
    print('Node {}, {}'.format(node1, node2))
    # print(df[df['node'] == node1]['letters'])
    # print(df[df['node'] == node2])

    voltage1 = float(df[df['node'] == node1]['voltage'].iloc[0])
    voltage2 = float(df[df['node'] == node2]['voltage'].iloc[0])
    voltage = voltage1 - voltage2
    # print(voltage)

    letters1 = df[df['node'] == node1]['letters'].iloc[0]
    letters2 = df[df['node'] == node2]['letters'].iloc[0]
    letters = letters1 + letters2
    # print(letters)

    data.append([node1, node2, voltage, letters])

for i in range(n):
    for j in range(n):
        
        # print('i, j: {}, {}'.format(i, j), n)
        
        # Horizontal resistors
        if j < 3:
            node1 = node_name(i, j)
            node2 = node_name(i, j+1)
            if node1 in df.values and node2 in df.values :
                each_step(node1, node2)

        # Vertical resistors
        if i < 3:
            node1 = node_name(i, j)
            node2 = node_name(i+1, j)        
            if node1 in df.values and node2 in df.values :
                each_step(node1, node2)
data
```

```python

df = pd.DataFrame(data, columns=['node1', 'node2', 'voltage', 'letters'])
df
```

```python
def swap_rows(df, row1, row2):
    df.iloc[row1], df.iloc[row2] =  df.iloc[row2].copy(), df.iloc[row1].copy()

swap_rows(df, 1, 2)
swap_rows(df, 2, 3)
swap_rows(df, 3, 4)
swap_rows(df, 4, 6)
swap_rows(df, 5, 8)
swap_rows(df, 6, 11)
swap_rows(df, 7, 10)
swap_rows(df, 8, 9)
swap_rows(df, 9, 10)
```

```python
# Find the midpoint of the DataFrame
midpoint = len(df) // 2

# Multiply the voltage column for the second half of the DataFrame by -1
df.loc[midpoint:, 'voltage'] *= -1
```

```python
# Swap the values of node1 and node2 for the second half of the DataFrame
df.loc[midpoint:, ['node1', 'node2']] = df.loc[midpoint:, ['node2', 'node1']].values

df
```

```python
# Reverse the letters for the second half of the DataFrame
df.loc[midpoint:, 'letters'] = df.loc[midpoint:, 'letters'].apply(lambda x: x[::-1])
df
```
