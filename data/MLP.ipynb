```python
from sklearn import neural_network
```

```python
model = neural_network.MLPRegressor(hidden_layer_sizes=(24, 30), max_iter=100000, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
# y_pred
```

```python
y = model.predict(X_train)
# y
```

```python
# y_test
```

```python
# tests MSE
mse = metrics.mean_squared_error(y_test, y_pred)
print("test MSE: ", mse)
```

```python
# train MSE
mse = metrics.mean_squared_error(y_train, y)
print("train MSE: ", mse)
```
